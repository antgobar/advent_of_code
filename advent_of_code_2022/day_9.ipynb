```python
data = """R 4
U 4
L 3
D 1
R 4
D 1
L 5
R 2""".split("\n")
```

```python
import math

with open("day_9_input.txt") as f:
    data = f.read().split("\n")

instructions = [tuple([d.split(" ")[0], int(d.split(" ")[1])]) for d in data]

def direction_parser(direction):
    direction_map = {
        "R": [1, 0],
        "L": [-1, 0],
        "U": [0, 1],
        "D": [0, -1]
    }
    return direction_map[direction]

def update_position(position, direction):
    return [p+d for p, d in zip(position, direction)]

def position_difference(head, tails):
    return [p-d for p, d in zip(head, tails)]

def pythagit(x, y):
    return math.sqrt(x**2 + y**2)

def stringify_coordinates(x, y):
    return f"{x}-{y}"

path_taken = ["0-0"]
head_position = [0, 0]
tail_position = [0, 0]
for instruction in instructions:
    step_direction, magnitude = instruction
    for _ in range(magnitude):
        direction = direction_parser(step_direction)
        new_head_position = update_position(head_position, direction)
        head_tail_difference = position_difference(new_head_position, tail_position)
        if pythagit(*head_tail_difference) > math.sqrt(2):
            tail_position = head_position
            path_taken.append(stringify_coordinates(*tail_position))
        head_position = new_head_position
    
print(len(set(path_taken)))
```
